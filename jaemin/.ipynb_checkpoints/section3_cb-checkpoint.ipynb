```python
ratings = pd.read_csv('ratings_p.csv')
genres = pd.read_csv('genres_d.csv')
```

```python
genres.head()
```

가중평균 활용

```python
# ratings에서 1번 유저의 데이터만 따로 떼어내 테이블 'user1' 만들기
user1 = ratings[ratings['userId']==1]
user1.head()
```

```python
# user1의 인덱스를 movieid로 설정
user1 = user1.set_index('movieId')
user1.head()
```

```python
# genres의 인덱스를 movieid로 설정
genres = genres.set_index('movieId')
genres.head()
```

```python
# genres에서 1번 유저가 평점을 준 영화의 데이터만 떼어내 user1_rated 테이블 만들기
user1_rated = genres[genres.index.isin(user1.index)]
user1_rated.head()
```

```python
# genres에서 1번 유저가 평점을 주지 않은 영화, 즉 추천작의 후보가 될 수 있는 데이터만 떼어내 user1_unrated 테이블 만들기
user1_unrated = genres[~genres.index.isin(user1.index)]
user1_unrated.head()
```

```python
print(user1_rated.shape)
print(user1_unrated.shape)
```

```python
# user1에서 평점만 가져오기
user1['rating']
```

```python
# user1['rating']과의 행렬 곱을 위해 user1_rated의 전치행렬(transposed matrix)을 만든다.
user1_movies = user1_rated.T
user1_movies
```

```python
# 1번 유저의 유저 프로필 = 장르별 가중평균 (weights average)
user1_profile = user1_movies.dot(user1['rating']) / sum(user1['rating'])
user1_profile
```

```python
# user1_unrated(아직 평가하지 않은 영화 테이블)와 유저프로필의 행렬 곱
user1_v = user1_unrated.dot(user1_profile).reset_index().rename(columns={0:'value'})
```

```python
user1_v
```

```python
titles = ratings[['movieId', 'title', 'genres']].drop_duplicates()
titles.shape
```

```python
user1_recommendations = user1_v.merge(titles, how='left', on='movieId').sort_values('value', ascending=False)
user1_recommendations.head(20)
```

cosine similarity 활용

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
genres.head()
```

```python
genres.loc[0:2]
```

```python
cosine_similarity(genres.loc[0:2])
```

```python
cos_sim = cosine_similarity(genres)
cos_sim
#    1번과 1번,    1번과 2번, ...    1번과 9742번
#    2번과 1번,    2번과 2번, ...    2번과 9742번
# ...
# 9742번과 1번, 9742번과 2번, ... 9742번과 9742번
```

```python
cos_sim.shape
```

```python
# 유사도가 작은 순으로 인덱스 출력
np.argsort(cos_sim[0])
```

```python
# 유사도가 큰 순으로 인덱스 출력
np.argsort(cos_sim[0])[::-1]
```

```python
top20 = np.argsort(cos_sim[0])[::-1][1:21]
top20
```

```python
movies = pd.read_csv('../../movies.csv')
tags = pd.read_csv('../../tags.csv')
```

```python
movies[movies.index.isin(top20)]
```
