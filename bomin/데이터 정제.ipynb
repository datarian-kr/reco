엄밀히 따져보자면...

1987이라는 이름의 영화가 있다고 해보자. 영화 개봉 년도가 따로 써있지 않고 title에 1987이라고만 써있다면 unique값을 보고 에러인지 알아낼 수 있을까?

아니라면 어떻게 해야 정확하게 년도 정보를 얻어올 수 있을까?

```python
# pandas str.extract 함수 이용
# Series 데이터를 넣고 정규표현식을 이용하여 분리해서 추출할 문자집합을 설정하면
# 각각의 문자집합들을 각각 컬럼으로 하나씩 뽑아준다.

# 여기서는 괄호에 둘러싸인 0개 이상의 숫자를 가져오라는 정규표현식을 설정함

movies['title'].str.extract(r'(\()(\d*)(\))')
```

```python
movies['year'] = movies['title'].str.extract(r'(\()(\d*)(\))')[1]
```

```python
movies
```

```python
movies['year'].unique()
```

```python
genres = list(movies['genres'].apply(lambda x: x.split('|')))
```

```python
genres_unique = list(set([g for genre in genres for g in genre]))
print(genres_unique)
```

```python
for g in genres_unique:
    movies[g] = movies['genres'].apply(lambda x: g in x)
```

```python
# 우리 데이터를 기준으로 가장 많이 개봉된 장르는 Drama, Comedy, Thriller 순
movies.sum()[genres_unique].sort_values(ascending=False)
```

```python
# 장르없는 영화들 영화 자체로만 보면 34개
movies[movies['(no genres listed)'] == True]
```

```python
# 평점이 많으면 채워주려고 했더니 평점 붙여봐도 총 평점 데이터 수가 47개 밖에 안됨.
# 한 영화에 리뷰 한두개 정도 있다는 것인데 그럼 그닥 중요하지 않은 데이터 인 것 같음.
# 설마 이런 영화들만 골라서 보는 사람 없겠지?

movies[movies['(no genres listed)'] == True].merge(ratings, on='movieId')
```

```python
# 111번과 599번 user가 그나마 많이 봤네
# 몇사람 제외하면 나머지는 거의 한 두개니까 한 사람당 리뷰 한두개 정도 빠지는 건 괜찮을 듯.
movies[movies['(no genres listed)'] == True].merge(ratings, on='movieId').groupby('userId')['movieId'].count()
```
